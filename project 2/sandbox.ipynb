```python
def loss_func(x,y,weights):

    # s=0
    # print(x.shape,y.shape,weights.shape)
    # for i in range(len(x)):
    #     s+=np.log(1+np.exp(-y[i].reshape([-1,1])*(x[i].dot(weights))))

    # h = 1/(1+np.exp(-x.dot(weights)))
    # log =  -np.sum((y*np.log(h)) + ((1-y)*np.log(h)))/len(y)
    # print(log)
    
    return np.sum(np.log(1+np.exp(-y*(x.dot(weights)))))
    
def grad_update(x,y, weights, feats_selection=[], learn_rate = 0.001, topk=1):
    assert weights.shape==(nfeats+1,1)

    # gradient = 0
    # for i in range(len(x)):
    #     # print(x.shape)
    #     gradient+= (y[i]*x[i])/(1+np.exp(y[i]*weights.T.dot(x[i])))

    gradient = np.sum((y*x)/(1+np.exp(y*(x.dot(weights)))),0).reshape([-1,1])
    
    # h = 1/(1+np.exp(-x.dot(weights)))
    # gradient = (x.T.dot(h-y))/len(y)

    if feats_selection=='random':
        gradient = np.random.choice(2,len(gradient)).reshape([-1,1])*gradient
    elif feats_selection=='top':
        t=gradient[gradient.argsort()[::-1][topk-1]].reshape(-1)
        gradient[gradient<t]=0
        # gradient[gradient!=np.max(gradient)]=0
        
        
    return weights + (learn_rate*gradient.reshape([-1,1]))
```

```python
runs = 1
max_iters = 1000
nfeats = len(X.columns)
topk=3

fig = go.Figure()
fin_loss = []

for i in range(runs):
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
    
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train=scaler.transform(X_train)
    X_test=scaler.transform(X_test)
    
    X_train = np.append(X_train,np.ones([len(X_train),1]),1)
    X_test = np.append(X_test,np.ones([len(X_test),1]),1)

    y_train = y_train.reshape([-1,1])
    y_test = y_test.reshape([-1,1])

    weights = np.random.random([nfeats+1,1])
    weightsRandomUpdate = weights.copy() 
    weightsTopGradUpdate = weights.copy() 

    loss = []
    lossTopGradUpdate = []
    lossRandomUpdate = []

    for it in range(max_iters):
        # All weights update
        weights = grad_update(X_train,y_train,weights)
        curr_loss = loss_func(X_train,y_train,weights)
        
        # Random weights update
        weightsRandomUpdate = grad_update(X_train,y_train,weightsRandomUpdate,feats_selection = 'random')
        curr_loss_rand_update = loss_func(X_train,y_train,weightsRandomUpdate)

        # Only Highest Gradient weight update
        weightsTopGradUpdate = grad_update(X_train,y_train,weightsTopGradUpdate,feats_selection='top',topk=topk)
        curr_loss_top_grad = loss_func(X_train,y_train,weightsTopGradUpdate)
        
        if it%100==0:
            print(f"Iteration {it} :: Current training loss: {curr_loss:0.2f} :: {curr_loss_rand_update:0.2f} :: {curr_loss_top_grad:0.2f}")

        loss.append(curr_loss)
        lossRandomUpdate.append(curr_loss_rand_update)
        lossTopGradUpdate.append(curr_loss_top_grad)
        
    fin_loss.append(curr_loss)

    fig.add_trace(go.Scatter(x = list(range(max_iters)), y = loss, mode = 'lines', name = f"All Weights Update"))
    
    fig.add_trace(go.Scatter(x = list(range(max_iters)), y = lossRandomUpdate, mode = 'lines', name = f"Random Weights Update"))
    
    fig.add_trace(go.Scatter(x = list(range(max_iters)), y = lossTopGradUpdate, mode = 'lines', name = f"Top {topk} Gradient Weight Update"))

fig.update_layout(
    autosize=False,
    width=800,
    height=600,
    legend=dict(
    yanchor="top",
    y=0.98,
    xanchor="right",
    x=0.99
)
)

fig.update_xaxes(
        title_text = "Iterations",
        title_standoff = 5)

fig.update_yaxes(
        title_text = "Loss",
        title_standoff = 5)

fig.show()
```

```python
def loss_func(x,y,weights):
    
    # s=0
    # print(x.shape,y.shape,weights.shape)
    # for i in range(len(x)):
    #     s+=np.log(1+np.exp(-y[i].reshape([-1,1])*(x[i].dot(weights))))

    # Log-loss
    # h = 1/(1+np.exp(-x.dot(weights)))
    # log =  -np.sum((y*np.log(h)) + ((1-y)*np.log(h)))/len(y)
    # print(log)
    
    return np.sum(np.log(1+np.exp(-y*(x.dot(weights)))))
   
def grad_update(x,y, weights, feats_selection=[], mode = 'all',learn_rate = 0.1, topk=1,curr_coord=0):
    assert weights.shape==(nfeats+1,1)
    if x.shape!=(nfeats+1,1) and mode!='all':
        x=x.reshape([-1,1])

    # gradient = 0
    # for i in range(len(x)):
    #     # print(x.shape)
    #     gradient+= (y[i]*x[i])/(1+np.exp(y[i]*weights.T.dot(x[i])))

    if mode=='all':
        gradient = -np.sum((y*x)/(1+np.exp(y*(x.dot(weights)))),0).reshape([-1,1])
    else:
        gradient = -(y*x)/(1+np.exp(y*(weights.T.dot(x)))).reshape([-1,1])
    
    # Log-loss
    # h = 1/(1+np.exp(-x.dot(weights)))
    # gradient = (x.T.dot(h-y))/len(y)
    

    if feats_selection=='random':
        # gradient = np.random.choice(2,len(gradient)).reshape([-1,1])*gradient
        temp = np.zeros(gradient.shape)
        random_coord = np.random.choice(14)
        temp[random_coord] = gradient[random_coord]
        gradient = temp

    elif feats_selection=='top':
        t=np.argmax(abs(gradient))
        gradient[gradient!=gradient[t]]=0
        
    elif feats_selection == 'round_robin':
        temp = np.zeros(gradient.shape)
        temp[curr_coord] = gradient[curr_coord]
        gradient = temp
        # gradient[gradient!=np.max(gradient)]=0
        
        
    return weights - (learn_rate*gradient.reshape([-1,1]))
```

```python
runs = 100
nfeats = len(X.columns)
topk=1
epochs = 500

fig = go.Figure()
fin_loss = []
# fin_test_loss = []

for i in range(runs):
    curr_coord = 1
    fig = go.Figure()
    
    # X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
    X_train,y_train = X.copy(),Y.copy()

    
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train=scaler.transform(X_train)
    # X_test=scaler.transform(X_test)
    
    X_train = np.append(X_train,np.ones([len(X_train),1]),1)
    # X_test = np.append(X_test,np.ones([len(X_test),1]),1)

    y_train = y_train.reshape([-1,1])
    # y_test = y_test.reshape([-1,1])

    weights = np.random.random([nfeats+1,1])
    weightsRandomUpdate = weights.copy() 
    weightsTopGradUpdate = weights.copy() 
    weightsRR = weights.copy() 

    loss = []
    lossTopGradUpdate = []
    lossRandomUpdate = []
    lossRR = []
    
    for e in range(epochs):        
        for it in range(len(X_train)):
            
            # All weights update            
            weights = grad_update(X_train[it],y_train[it],weights,mode='stochastic')
            curr_loss = loss_func(X_train,y_train,weights)
            
            # Random weights update
            weightsRandomUpdate = grad_update(X_train[it],y_train[it],weightsRandomUpdate,feats_selection = 'random',mode='stochastic')
            curr_loss_rand_update = loss_func(X_train,y_train,weightsRandomUpdate)
            
            # Only Highest Gradient weight update
            weightsTopGradUpdate = grad_update(X_train[it],y_train[it],weightsTopGradUpdate,feats_selection='top',topk=topk,mode='stochastic')
            curr_loss_top_grad = loss_func(X_train,y_train,weightsTopGradUpdate) 
            
            # Round Robin gradient update
            weightsRR = grad_update(X_train[it],y_train[it],weightsRR,mode='stochastic',feats_selection='round_robin',curr_coord=curr_coord)
            curr_loss_rr = loss_func(X_train,y_train,weightsRR)
        
        curr_coord = (curr_coord+1)%(nfeats+1)  
            
        loss.append(curr_loss)
        lossRandomUpdate.append(curr_loss_rand_update)
        lossTopGradUpdate.append(curr_loss_top_grad)
        lossRR.append(curr_loss_rr)

        if e%500==0:
                print(f"Epoch {e} :: Current training loss: {curr_loss:0.2f} :: {curr_loss_rand_update:0.2f} :: {curr_loss_top_grad:0.2f} :: {curr_loss_rr:0.2f}")   
        
    fin_loss.append((curr_loss,curr_loss_rand_update,curr_loss_top_grad,curr_loss_rr))
    # fin_test_loss.append((loss_func(X_test,y_test,weights),loss_func(X_test,y_test,weightsRandomUpdate),loss_func(X_test,y_test,weightsTopGradUpdate),loss_func(X_test,y_test,weightsRR)))

    fig.add_trace(go.Scatter(x = list(range(epochs)), y = loss, mode = 'lines', name = f"All Weights Update"))
    
    fig.add_trace(go.Scatter(x = list(range(epochs)), y = lossRandomUpdate, mode = 'lines', name = f"Random Weights Update"))
    
    fig.add_trace(go.Scatter(x = list(range(epochs)), y = lossTopGradUpdate, mode = 'lines', name = f"Max Gradient Weight Update"))

    fig.add_trace(go.Scatter(x = list(range(epochs)), y = lossRR, mode = 'lines', name = f"Round Robin Weight Update"))

fig.update_layout(
    autosize=False,
    width=800,
    height=600,
    legend=dict(
    yanchor="top",
    y=0.98,
    xanchor="right",
    x=0.99
)
)

fig.update_xaxes(
        title_text = "Epoch",
        title_standoff = 5)

fig.update_yaxes(
        title_text = "Loss",
        title_standoff = 5)

fig.show()
```
